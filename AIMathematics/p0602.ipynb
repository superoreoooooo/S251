```python
import sympy as sym

x = sym.Symbol('x')
f = x**3-1
g = x**2-1
val = 1

lim_f = sym.limit(f, x, val)
lim_g = sym.limit(g, x, val)

diff_f = sym.diff(f, x)
diff_g = sym.diff(g, x)

LHospital_val = sym.limit(diff_f/diff_g, x, val)

print(LHospital_val)
```

```python
import sympy as sym

# lamdify 함수 : sympy 표현식을 파이썬 함수로 변환
def cal_critical_point(f, x):
    diff_func = sym.diff(f, x)
    critical_point = sym.solve(diff_func, x)
    return critical_point

def second_diff_test(f, critical_point):
    diff_f_x = sym.diff(f, x)
    diff_f_2 = sym.diff(diff_f_x, x)
    origin_f_func = sym.lambdify(x, f, 'numpy')
    diff_func = sym.lambdify(x, diff_f_x, 'numpy')
    diff_func_2 = sym.lambdify(x, diff_f_2, 'numpy')
    
    check_val = diff_func_2(critical_point)
    origin_val = origin_f_func(critical_point)
    critical_val = diff_func(critical_point)

    point_info = f"({critical_point}, {check_val})"
    return point_info, origin_val

x = sym.Symbol('x')
f = (1/4)*x**4 - x**3 + x**2 + 1

diff_f = sym.diff(f, x)
diff_f_2 = sym.diff(diff_f, x)
critical_point_list = cal_critical_point(f, x)

cri_info_1, cri_val_1 = second_diff_test(f, critical_point_list[0])
cri_info_2, cri_val_2 = second_diff_test(f, critical_point_list[1])
cri_info_3, cri_val_3 = second_diff_test(f, critical_point_list[2])

print("f의 도함수 =", diff_f)
print("f의 이계도함수 =", diff_f_2)
print()
print("f의 극값점 =", critical_point_list)
print()
print("f의 극값 정보 1 :", cri_info_1, "값 :", cri_val_1)
print("f의 극값 정보 2 :", cri_info_2, "값 :", cri_val_2)
print("f의 극값 정보 3 :", cri_info_3, "값 :", cri_val_3)

# 함수 f(x)의 선 그래프 그리기
import matplotlib.pyplot as plt
import numpy as np

my_f_func = sym.lambdify(x, f, 'numpy')
x = np.arange(-2, 4, 0.1)

# 라벨이름 넣기
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.grid(color = 'black', alpha=0.5, linestyle='--')

# 그래프 그리기
plt.plot(x, my_f_func(x), label='y = (1/4)x^4 - x^3 + x^2 + 1')

# 범례
plt.legend()
print("함수 f(x)의 선 그래프")
plt.show()
```
