```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-2, 8, 0.1)
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.grid(color="black", alpha=0.5, linestyle="--")

plt.plot(x, np.exp(x))

plt.legend()
plt.show()
```

```python
import sympy as sym

h = sym.Symbol('h')
f = (((1+h)**3)-(1**3))/h
lim_f = sym.limit(f, h, 0)
print(lim_f)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = sym.Symbol('x')
f = 1/3*(x**3) + 1/2*(x**2) + x + 1

diff_f = sym.diff(f, x)
diff2_f = sym.diff(diff_f, x)

print(diff_f, diff2_f)

x = np.arange(-5, 5, 0.1)
plt.plot(x, 1/3*(x**3) + 1/2*(x**2) + x + 1)
plt.grid()
plt.show()
```

```python
t = sym.Symbol('t')
s = t**3 + t

diff_t = s.diff()
diff2_t = diff_t.diff()

print(diff_t, diff2_t)

t = np.arange(-5, 5, 0.1)

plt.plot(t, t**3 + t)
plt.grid()
plt.show()
```

```python
import sympy as sym

#  (a)
x = sym.Symbol('x')
h = (x+3)**2 + 20
diff_h = sym.diff(h, x)
print("(a) h  =", diff_h)

#  (b) :  
x = sym.Symbol('x')
h = sym.cos(x) * (x**2)
diff_h = sym.diff(h, x)
print("(b) h  =", diff_h)

#  (c)
x = sym.Symbol('x')
h = sym.exp(x**2 + 1)
diff_h = sym.diff(h, x)
print("(c) h  =", diff_h)
```
